# Determine Shipping Routes

```python
import pandas as pd
import numpy as np
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')
```

# Order Region + Customer Country + Shipping Mode

```python
df['detailed_route'] = (
    df['Order Region'].astype(str) + ' -> ' +
    df['Customer Country'].astype(str) + ' | ' +
    df['Shipping Mode'].astype(str)
)

detailed_routes = (
    df.groupby('Order Region')['detailed_route']
      .nunique()
      .reset_index()
)

detailed_routes.columns = [
    'Order Region',
    'Unique Detailed Routes'
]

print(detailed_routes)
```

```python
routes_list = (
    df.groupby('Order Region')['shipping_route']
      .unique()
)

print(routes_list)
```
